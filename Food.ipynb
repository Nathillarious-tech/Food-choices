```python
# Convert numeric-like columns
df['GPA'] = pd.to_numeric(df['GPA'], errors='coerce')
df['calories_day'] = pd.to_numeric(df['calories_day'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')
df['fruit_day'] = pd.to_numeric(df['fruit_day'], errors='coerce')
df
```

```python
df.fillna({
    'calories_day':df['calories_day'].median(),
    'calories_scone':df['calories_scone'].median(),
    'comfort_food':df['comfort_food'].mode()[0],
    'comfort_food_reasons':df['comfort_food_reasons'].mode()[0],
    'comfort_food_reasons_coded':df['comfort_food_reasons_coded'].mean(),
    'cook':df['cook'].median(),
    'cuisine':df['cuisine'].mean(),
    'drink':df['drink'].median(),
    'diet_current':df['diet_current'].mode()[0],
    'eating_changes':df['eating_changes'].mode()[0],
    'employment':df['employment'].median(),
    'exercise':df['exercise'].median(),
    'father_education':df['father_education'].median(),
    'father_profession':df['father_profession'].mode()[0],
    'fav_food':df['fav_food'].median(),
    'income':df['income'].median(),
    'fav_cuisine':df['fav_cuisine'].mode()[0],
    'food_childhood':df['food_childhood'].mode()[0],
    'GPA':df['GPA'].mean(),
    'healthy_meal':df['healthy_meal'].mode()[0],
    'ideal_diet':df['ideal_diet'].mode()[0],
    'life_rewarding':df['life_rewarding'].median(),
    'marital_status':df['marital_status'].median(),
    'meals_dinner_friend':df['meals_dinner_friend'].mode()[0],
    'mother_education':df['mother_education'].median(),
    'mother_profession':df['mother_profession'].mode()[0],
    'on_off_campus':df['on_off_campus'].median(),
    'persian_food':df['persian_food'].median(),
    'self_perception_weight':df['self_perception_weight'].median(),
    'soup':df['soup'].median(),
    'sports':df['sports'].median(),
    'tortilla_calories':df['tortilla_calories'].mean(),
    'type_sports':df['type_sports'].mode()[0],
    'weight':df['weight'].median(),
    
}, inplace=True)
```

```python
df['Gender'] = df['Gender'].map({1: "Female", 2: "Male"})
df
```

```python
#normalize text
text_columns = [
    'comfort_food',
    'comfort_food_reasons',
    'eating_changes',
    'diet_current',
    'father_profession',
    'fav_cuisine',
    'food_childhood',
    'healthy_meal',
    'ideal_diet',
    'meals_dinner_friend',
    'mother_profession',
    'parents_cook',
    'type_sports'
]
for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.title()

df
```

```python
#converting numeric to float
numeric_columns = ['GPA', 'calories_day', 'weight', 'fruit_day',
                   'calories_chicken', 'calories_scone']

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df
```

```python
df.info()
```

```python
# Remove invalid GPA and calorie values
df = df[(df['GPA'] >= 0) & (df['GPA'] <= 4)]
df = df[(df['calories_scone'] >= 0) & (df['calories_scone'] < 10000)]
df
```

```python
def bmi_category(weight):
    if pd.isnull(weight):
        return "Unknown"
    elif weight < 100:
        return "Underweight"
    elif 100 <= weight <= 180:
        return "Normal"
    else:
        return "Overweight"

df['BMI_Category'] = df['weight'].apply(bmi_category)
df['BMI_Category']
```

```python
#Group by Gender and calculate averages
gender_group = df.groupby('Gender')[['GPA', 'fruit_day', 'calories_day']].mean()
print(gender_group)
```

```python
# Create a new column comparing each student's GPA to the average
avg_gpa = df['GPA'].mean()

df['GPA_Category'] = df['GPA'].apply(
    lambda x: "Above Avg" if x > avg_gpa else "Below Avg"
)
print("Average GPA:", avg_gpa)
print(df[['GPA', 'GPA_Category']].head(10))
```

```python
#commmon comfort food
comfort_food = df['comfort_food'].dropna()
comfort_foods_split = comfort_food.str.split(',').explode().str.strip().str.title()
comfort_food_counts = comfort_food.value_counts()
print(comfort_food_counts.head())

```

```python
#relationship between exercise frequency and GPA
exercise_gpa = df.groupby('exercise')['GPA'].mean()
print("Average GPA by exercise frequency:")
print(exercise_gpa)
```

```python
#Gender that consumes more coffee
coffee_by_gender = df.groupby('Gender')['coffee'].mean()
print("Average coffee consumption by gender:")
print(coffee_by_gender)
```
